```python
## Cambia el nombre de la columna.

df.rename(columns={'availability 365': 'days_booked'}, inplace=True)
df.columns
```

```python
## Convierte todos los nombres de columna a minúsculas y sustituye los espacios por un guión bajo "_".
df.columns = [col.lower().replace(" ", "_") for col in df.columns]
df.columns
```

```python
## Elimina el signo de dólares y la coma de las columnas. Si es necesario, convierte estas dos columnas al tipo de datos adecuado.

#Importamos numpy
import numpy as np
#Creamos una función a doc con lo aprendido para eliminar $ y ","
def remove_dollar_sign(value):
    if pd.isna(value):
        return np.NaN
    else:
        if isinstance(value, str):  # Comprueba si el valor es una cadena
            return float(value.replace("$", "").replace(",", "").replace(" ", ""))
        else:
            return value  # Si no es una cadena, devuelve el mismo valor

# Aplica la función remove_dollar_sign a la columna 'price' y de 'service_fee'
df['price'] = df['price'].apply(remove_dollar_sign)
df['service_fee'] = df['service_fee'].apply(remove_dollar_sign)
# Muestra el DataFrame resultante
df.price
```

```python
#Enumere los 10 barrios más caros por orden creciente de precio con la ayuda de un gráfico de barras horizontales. ¿Cuál es el barrio más barato?
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")

grp2=df['price'].groupby(df['neighbourhood']).sum().sort_values(ascending=False)
grp2.head()
ax = plt.gca()
ax.get_xaxis().get_major_formatter().set_scientific(False)
grp2.head(10).plot(kind='barh',color={'green'})
plt.xlabel("Price")
plt.ylabel("Neighborhood")

```

```python
#Enumere los barrios que ofrecen alquileres a corto plazo de menos de 10 días. Ilustrar con un gráfico de barras
# Convertir la columna 'minimum_nights' a tipo numérico (si está en formato de cadena)
df['minimum_nights'] = pd.to_numeric(df['minimum_nights'], errors='coerce')
short_term_rentals = df[df['minimum_nights'] < 10]
short_term_rentals
```

```python
# Aquí enumero la cantidad de alquileres a corto plazo por barrio
short_term_rentals_count = short_term_rentals['neighbourhood_group'].value_counts().head(10)
short_term_rentals_count
```

```python
# Voy a graficar los 10 primeros barrios con alquileres a corto plazo
plt.figure(figsize=(10, 6))
short_term_rentals_count.plot(kind='bar')
plt.title('Top 10 Barrios con Alquileres a Corto Plazo')
plt.xlabel('Barrios')
plt.ylabel('Cantidad de Alquileres a Corto Plazo')
plt.xticks(rotation=45)
plt.show()
```

```python
#Enumere los precios con respecto al tipo de habitación utilizando un gráfico de barras y exponga también sus inferencias.
# Calculo el precio promedio por tipo de habitación
price_by_room_type = df.groupby('room_type')['price'].mean().nlargest(7).sort_values(ascending=False)

# Graficar los precios por tipo de habitación en un gráfico de barras
plt.figure(figsize=(8, 6))
price_by_room_type.plot(kind='bar', color='skyblue')
plt.title('Precios por Tipo de Habitación')
plt.xlabel('Tipo de Habitación')
plt.ylabel('Precio Promedio')
plt.xticks(rotation=45)
plt.show()
```

```python
#Inferencias: El gráfico de barras generado te permite comparar visualmente los precios promedio entre los diferentes tipos de habitaciones. 
#Puedes observar cuál tipo de habitación tiende a ser más costoso en promedio y también identificar cualquier brecha significativa en los precios 
#entre los tipos de habitaciones. 
#Esto puede ser útil para comprender cómo varían los precios según el tipo de alojamiento ofrecido y tomar decisiones basadas en estos patrones, 
#como fijar precios competitivos o comprender la demanda de ciertos tipos de habitaciones en el mercado
```

```python
#Cree un gráfico circular que muestre la distribución de los días reservados para cada grupo de barrios.
# Convertir la columna 'days_booked' a tipo numérico (si está en formato de cadena)
df['days_booked'] = pd.to_numeric(df['days_booked'], errors='coerce')

# Calcular la suma de días reservados por grupo de barrios y tomar los 10 primeros
days_reserved_by_group = df.groupby('neighbourhood_group')['days_booked'].sum().nlargest(5)

# Graficar la distribución de días reservados en un gráfico circular (pie chart) para los 10 primeros grupos de barrios
plt.figure(figsize=(8, 6))
plt.pie(days_reserved_by_group, labels=days_reserved_by_group.index, autopct='%1.1f%%', startangle=140)
plt.legend(title='Grupos de Barrios', loc='upper right')
plt.title('Distribución de Días Reservados por Grupo de Barrios')
plt.subplots_adjust(bottom=0.2)  # Ajustar el margen inferior
plt.axis('equal')
plt.tight_layout()
plt.show()
```

```python
#Utilizando un gráfico lineal muestre en qué año tuvo lugar la máxima construcción de habitaciones.
# Convertir la columna 'construction_year' a tipo numérico (si está en formato de cadena)
df['construction_year'] = pd.to_numeric(df['construction_year'], errors='coerce')

# Contar la cantidad de habitaciones construidas por año
rooms_built_per_year = df['construction_year'].value_counts().sort_index()

# Graficar la cantidad de habitaciones construidas por año (gráfico lineal)
plt.figure(figsize=(10, 6))
rooms_built_per_year.plot(kind='line', marker='o', color='skyblue')
plt.title('Cantidad de Habitaciones Construidas por Año')
plt.xlabel('Año de Construcción')
plt.ylabel('Cantidad de Habitaciones Construidas')
plt.grid(True)

plt.show()
```
